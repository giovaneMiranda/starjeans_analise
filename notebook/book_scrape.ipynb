# Solução  

```python
# importing packages
from bs4 import BeautifulSoup
import requests  
import pandas as pd 
```

```python
def scrape_book(type_book):

    url = 'https://books.toscrape.com/catalogue/category/books/' + type_book + '/index.html'

    book_name = []
    book_rating = []
    book_price = []
    book_stock = []

    page = requests.get(url, header)
    soup = BeautifulSoup(page.text, 'html.parser')

    book_list = soup.find_all('li',
                              class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')

    for b in book_list:
        book_name.append(b.find('img', class_='thumbnail').get('alt'))

        book_rating.append(
            b.find('i', class_='icon-star').find_parent('p').get('class')[1])

        book_price.append(b.find('p', class_='price_color').get_text())

        book_stock.append('-'.join(
            b.find('p', class_='instock availability').get_text().split()))

    df_book = pd.DataFrame([book_name, book_price, book_stock, book_rating]).T
    df_book.columns = ['book_name', 'book_price', 'book_stock', 'book_rating']
    df_book['book_type'] = type_book
    return df_book
```

```python
header = {
    'User-Agent':
    'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'
}
```

```python
type_list = ['classics_6', 'science-fiction_16', 'humor_30', 'business_35']
df_book_raw = pd.DataFrame()
```

```python
df_book_raw = pd.concat([scrape_book(type_b) for type_b in type_list], axis = 0 ,ignore_index= True)
```
